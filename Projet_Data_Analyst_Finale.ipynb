```python
armes = test.armed
df3 = test.id
 
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
 
colors = ("orange", "cyan", "brown", "grey", "indigo", "beige")
 
wp = {'linewidth': 1, 'edgecolor': "green"}
 
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
 
fig, ax = plt.subplots(figsize=(18, 12))
wedges, texts, autotexts = ax.pie(df3, autopct=lambda pct: func(pct, df3), explode=explode, labels=armes,
                                  shadow=True, colors=colors, startangle=90, wedgeprops=wp,
                                  textprops=dict(color="magenta"))
 
ax.legend(wedges, armes, title="Armes", loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=8, weight="bold")
ax.set_title("Diffrents armes utiliss")
 
plt.show()
```
